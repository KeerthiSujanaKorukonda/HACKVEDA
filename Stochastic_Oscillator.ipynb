```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf

def stochastic_oscillator(df, period=14, smooth_k=3):
    # Calculate the %K line
    df['L14'] = df['Low'].rolling(window=period).min()
    df['H14'] = df['High'].rolling(window=period).max()
    df['%K'] = 100 * ((df['Close'] - df['L14']) / (df['H14'] - df['L14']))

    # Calculate the %D line
    df['%D'] = df['%K'].rolling(window=smooth_k).mean()

    return df

# Fetch historical stock data
ticker = 'AAPL'
data = yf.download(ticker, start='2024-04-01', end='2024-07-05')

# Calculate the Stochastic Oscillator
data = stochastic_oscillator(data)

# Plot the results
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 14), gridspec_kw={'height_ratios': [3, 1, 1]})

# Plot the candlestick chart with colored candles
mpf.plot(data, type='candle', ax=ax1, volume=ax2, show_nontrading=True, style='charles')

# Plot the Stochastic Oscillator
ax3.plot(data['%K'], label='%K', color='blue')
ax3.plot(data['%D'], label='%D', color='red')
ax3.axhline(y=80, color='r', linestyle='--')
ax3.axhline(y=20, color='g', linestyle='--')
ax3.set_title(f'{ticker} Stochastic Oscillator')
ax3.legend()

plt.tight_layout()
plt.show()
```
